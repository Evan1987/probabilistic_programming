## Bayesian Estimation Supersedes the T-test (B.E.S.T.)

```python
import numpy as np
import matplotlib.pyplot as plt
import pymc as pm
```

```python
N = 250

#  用户在 A、B页面停留时长的真实分布
params = {"A": (30, 4), "B": (26, 7)}
# 生成的真实样本（已观测）
durations = {k: np.random.normal(mu, std, size=N) for k, (mu, std) in params.items()}
```

**目标：判断用户在哪个页面停留更长**

```python
total = np.hstack(list(durations.values()))
mean_, std_ = total.mean(), total.std()

# 定先验分布
tau = 1.0 / (1000 * std_) ** 2
mu_A = pm.Normal("mu_A", mean_, tau)
mu_B = pm.Normal("mu_B", mean_, tau)

std_A = pm.Uniform("std_A", std_ / 1000, std_ * 1000)
std_B = pm.Uniform("std_B", std_ / 1000, std_ * 1000)

nu_minus_1 = pm.Exponential("nu-1", 1. / 29)

obs_A = pm.NoncentralT("obs_A", mu=mu_A, lam=1.0 / std_A ** 2, nu=nu_minus_1 + 1, 
                       observed=True, value=durations["A"])

obs_B = pm.NoncentralT("obs_B", mu=mu_B, lam=1.0 / std_B ** 2, nu=nu_minus_1 + 1, 
                       observed=True, value=durations["B"])

mcmc = pm.MCMC([obs_A, obs_B, mu_A, mu_B, std_A, std_B, nu_minus_1])
mcmc.sample(25000, 10000)
```

$$x_i \sim t(\mu,\sigma,\nu)$$  
$$\mu \sim N(M_{\mu}, S_{\mu})$$  
$$\sigma \sim U(L_{\sigma}, H_{\sigma})$$ 
$$\nu \sim ShiftedExp(\frac{1}{29}, shift=1)$$


```python
mu_A_trace, mu_B_trace = mcmc.trace("mu_A")[:], mcmc.trace("mu_B")[:]
std_A_trace, std_B_trace = mcmc.trace("std_A")[:], mcmc.trace("std_B")[:]
nu_trace = mcmc.trace("nu-1")[:] + 1
```

```python
plt.figure(figsize=(12, 8))

def _hist(data, label):
    plt.hist(data, bins=40, histtype="stepfilled", alpha=0.95, label=label, density=True)

# mu的后验分布
plt.subplot(311)
_hist(mu_A_trace, "A")
_hist(mu_B_trace, "B")
plt.legend()
plt.grid()
plt.title("Posterior distributions of $\mu$")

# sigma的后验分布
plt.subplot(312)
_hist(std_A_trace, "A")
_hist(std_B_trace, "B")
plt.legend()
plt.grid()
plt.title("Posterior distributions of $\sigma$")

# nu的后验分布
plt.subplot(313)
_hist(nu_trace, "")
plt.grid()
plt.title(r"Posterior distribution of $\nu$")

plt.xlabel("Value")
plt.ylabel("Density")
plt.tight_layout()
```
