```python
import numpy as np
import pymc as pm
import matplotlib.pyplot as plt
```

$$E[R]=79P_{79} + 49P_{49} + 25P_{25} + 0P_0$$  
where $$P_{79} + P_{49} + P_{25} + P_0 = 1$$

```python
N = 1000
N_79, N_49, N_25 = 10, 46, 80
N_0 = N - N_79 - N_49 - N_25

obs = np.array([N_79, N_49, N_25, N_0])
prior_params = np.ones(len(obs))  # [1, 1, 1, 1]
posterior_samples = np.random.dirichlet(prior_params + obs, size=10000)

for i in range(2):
    print(posterior_samples[i])
```

为后验绘制概率密度函数

```python
plt.figure(figsize=(12.5, 4))

for i, label in enumerate(["P_79", "P_49", "P_25", "P_0"]):
    plt.hist(posterior_samples[:, i], density=True, histtype="stepfilled", bins=50, label=label)
plt.xlabel("Prob")
plt.ylabel("Density")
plt.title("Posterior distributions of probability for selecting different prices")
plt.legend()
plt.grid()
```

回报期望

```python
posterior_expected_revenue = np.dot(posterior_samples, np.array([79, 49, 25, 0]))

plt.figure(figsize=(12.5, 4))
plt.hist(posterior_expected_revenue, histtype="stepfilled", density=True, label="expected revenue", bins=50)
plt.xlabel("Revenue")
plt.ylabel("Density")
plt.title("Posterior distributions of expected revenue")
plt.legend()
plt.grid()
```

## A/B测试

```python
data = {"A": [10, 46, 80, 864], "B": [45, 84, 200, 1671]}

def expected_posterior_revenue(obs):
    obs = np.array(obs)
    prior_params = np.ones_like(obs)
    distr = np.random.dirichlet(prior_params + obs, size=10000)
    
    prices = np.array([79, 49, 25, 0])
    return np.dot(distr, prices)


plt.figure(figsize=(12.5, 4))


expected_revenues = {}
for i, (k, obs) in enumerate(data.items()):
    expected_value = expected_posterior_revenue(obs)
    expected_revenues[k] = expected_value
    plt.hist(expected_value, histtype="stepfilled", density=True, 
             label=f"expected revenue of {k}", bins=50, alpha=1 - 0.2 * i)
plt.xlabel("Revenue")
plt.ylabel("Density")
plt.grid()
plt.title("Posterior distributions of the expected revenue between pages $A$ and $B$")
plt.legend()
```

```python
diff = expected_revenues["B"] - expected_revenues["A"]
print("Probability that page B  > page A", (diff > 0).mean())

plt.figure(figsize=(12.5, 4))
plt.hist(diff, histtype="stepfilled", color="#7A68A6", label="difference in revenue between B and A", bins=50)
plt.vlines(0, 0, 700, linestyles="--")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Posterior distribution of the delta between $B$ and $A$")
plt.legend()
plt.grid()
```
