```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
data = {"A": (1300, 120), "B": (1275, 125)}

alpha_prior, beta_prior = 1, 1
posteriors = {k: ss.beta(alpha_prior + conversions, beta_prior + visitors - conversions) 
              for k, (visitors, conversions) in data.items()}
```

```python
samples = 2000
samples_posterior = {k: distr.rvs(samples) for k, distr in posteriors.items()}
```

```python
print("A better than B: ", (samples_posterior["A"] > samples_posterior["B"]).mean())
```

```python
plt.figure(figsize=(12.5, 4))

x = np.linspace(0, 1, 500)
for k, distr in posteriors.items():
    plt.plot(x, distr.pdf(x), label=f"Posterior of {k}")
plt.xlim(0.05, 0.15)
plt.xlabel("Prob")
plt.ylabel("Density")
plt.grid()
plt.legend()
```

>**7.5 增幅问题**

```python
def relative_increase(a, b):
    return (b - a) / a
```

```python
posterior_relative_increase = relative_increase(samples_posterior["A"], samples_posterior["B"])

plt.figure(figsize=(12.5, 4))

plt.hist(posterior_relative_increase, histtype="stepfilled", bins=40, alpha=0.95, density=True)
plt.xlabel("Increase Ratio")
plt.ylabel("Density")
plt.title("Posterior distributions of relative lift of B over A")
plt.grid()
```
