```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as ss
```

```python
dist = ss.beta
data = iter(ss.bernoulli.rvs(0.5, size=2000))
n_trials = [0, 1, 2, 3, 4, 5, 8, 16, 50, 100, 500, 1000]
fig, axes = plt.subplots(nrows=len(n_trials) // 2, ncols=2, figsize=(11, 9))
x = np.linspace(0, 1, 100)

count = 0
heads = 0
fig_index = 0
while n_trials:
    if count == n_trials[0]:
        axe = axes[fig_index // 2, fig_index % 2]
        y = dist.pdf(x, 1 + heads, 1 + count - heads)
        axe.plot(x, y, label=f"{count} tosses, {heads} heads")
        axe.fill_between(x, 0, y, color="#348ABD", alpha=0.5)
        axe.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)
        axe.legend()
        axe.autoscale(tight=True)
        n_trials.pop(0)
        fig_index += 1
    
    call = next(data)
    heads += call
    count += 1
        
fig.suptitle("Bayesian updating of posterior probabilities", y=1.02, fontsize=14)
fig.tight_layout()
fig.show()
```
