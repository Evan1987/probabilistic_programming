### 后验与真实值

```python
import pymc as pm
import matplotlib.pyplot as plt

p = pm.Uniform("p", lower=0, upper=1)
p_true = 0.05  # unknown 
N = 1500

occurrences = pm.rbernoulli(p_true, N)
print("Observed freq: ", occurrences.mean())

obs = pm.Bernoulli("obs", p=p, value=occurrences, observed=True)
mcmc = pm.MCMC([obs, p])
mcmc.sample(20000, 1000)

plt.figure(figsize=(12.5, 4))
plt.title("Posterior distribution of $P_A$ (true ctr)")
plt.hist(mcmc.trace("p")[:], bins=35, histtype="stepfilled", density=True)
plt.vlines(p_true, 0, 90, linestyle="--", label="the true $P_A$(unknown)")
plt.xlabel("Value of $P_A$")
plt.ylabel("Density")
plt.legend()
```

### AB测试

```python
p_true_a = 0.05
p_true_b = 0.04
N_a = 1500
N_b = 750

observations_a = pm.rbernoulli(p_true_a, N_a)
observations_b = pm.rbernoulli(p_true_b, N_b)
print("Observation A mean: ", observations_a.mean())
print("Observation B mean: ", observations_b.mean())
```

```python
p_a = pm.Uniform("p_a", lower=0, upper=1)
p_b = pm.Uniform("p_b", lower=0, upper=1)

@pm.deterministic
def delta(p_a=p_a, p_b=p_b):
    return p_a - p_b

obs_a = pm.Bernoulli("obs_a", p_a, value=observations_a, observed=True)
obs_b = pm.Bernoulli("obs_b", p_b, value=observations_b, observed=True)

mcmc = pm.MCMC([obs_a, obs_b, delta, p_a, p_b])
mcmc.sample(25000, 5000)

p_a_samples = mcmc.trace("p_a")[:]
p_b_samples = mcmc.trace("p_b")[:]
delta_samples = mcmc.trace("delta")[:]

plt.figure(figsize=(12.5, 10))
ax = plt.subplot(311)
plt.xlim(0, 0.1)
plt.hist(p_a_samples, histtype="stepfilled", bins=30, alpha=0.85, 
         label="Posterior of $P_A$", color="#A60628", density=True)
plt.vlines(p_true_a, 0, 80, linestyles="--", label="the true $P_A$(unknown)")
plt.legend(loc="upper left")

ax = plt.subplot(312)
plt.xlim(0, 0.1)
plt.hist(p_b_samples, histtype="stepfilled", bins=30, alpha=0.85, 
         label="Posterior of $P_B$", color="#467821", density=True)
plt.vlines(p_true_b, 0, 80, linestyles="--", label="the true $P_B$(unknown)")
plt.legend(loc="upper left")

ax = plt.subplot(313)
plt.hist(delta_samples, histtype="stepfilled", bins=30, alpha=0.85, 
         label="Posterior of delta", color="#7A68A6", density=True)
plt.vlines(p_true_a - p_true_b, 0, 60, linestyles="--", label="the true delta(unknown)")
plt.legend(loc="upper left")

plt.ylabel("Density")
plt.xlabel("Value")
plt.title("The Posterior distribution of $P_A$,$P_B$ and delta unknowns")
```

```python
plt.figure(figsize=(12.5, 3))
plt.hist(p_a_samples, histtype="stepfilled", bins=30, alpha=0.6, 
         label="Posterior of $P_A$", color="#A60628", density=True)
plt.hist(p_b_samples, histtype="stepfilled", bins=30, alpha=0.6, 
         label="Posterior of $P_B$", color="#467821", density=True)
plt.legend(loc="upper right")
plt.xlim(0, 0.1)
plt.ylim(0, 80)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Posterior distributions of $P_A$ and $P_B$")
```

### 推断错误的概率

```python
print("The Prob. that A is worse than B: ", (delta_samples < 0).mean())
print("The Prob. that A is better than B: ", (delta_samples > 0).mean())
```
