```python
import pymc as pm
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 4
hyper_param = np.eye(n)

plt.figure(figsize=(20, 4))
for i in range(5):
    ax = plt.subplot(1, 5, i + 1)
    plt.imshow(pm.rwishart(n + 1, hyper_param), interpolation="none", cmap="hot")
    ax.axis("off")
```

```python
n = 15
hyper_param = np.eye(n) * 10

plt.figure(figsize=(20, 4))
for i in range(5):
    ax = plt.subplot(1, 5, i + 1)
    plt.imshow(pm.rwishart(n + 1, hyper_param), interpolation="none", cmap="hot")
    ax.axis("off")
```
