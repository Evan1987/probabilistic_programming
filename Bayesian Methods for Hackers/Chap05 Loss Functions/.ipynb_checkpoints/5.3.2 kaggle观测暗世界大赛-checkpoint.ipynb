```python
import numpy as np
import pandas as pd
import pymc as pm
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse


true_halos = pd.read_csv("data/Training_halos.csv")
true_halos[:5]
```

```python
def load_sky(sky_no: int):
    galaxy_data = pd.read_csv(f"data/Train_Skies/Training_Sky{sky_no}.csv")
    galaxy_data["d"] = np.sqrt(galaxy_data["e1"] ** 2 + galaxy_data["e2"] ** 2)
    galaxy_data["a"] = 1 / (1 - galaxy_data["d"])
    galaxy_data["b"] = 1 / (1 + galaxy_data["d"])
    galaxy_data["theta"] = np.degrees(np.arctan2(galaxy_data["e2"], galaxy_data["e1"]) * 0.5)
    return galaxy_data
```

```python
sky_no = 3
data = load_sky(sky_no)
data[:5]
```

```python
def draw_sky(galaxy_data: pd.DataFrame):
    
    fig, ax = plt.subplots(figsize=(10, 10))
    SIZE = 45

    for _, row in galaxy_data.iterrows():
        patch = Ellipse(xy=(row["x"], row["y"]), width=SIZE * row["a"], 
                        height=SIZE * row["b"], angle=row["theta"])
        ax.add_patch(patch)
    
    ax.autoscale_view(tight=True)
    ax.grid()
    return ax
```

```python
ax = draw_sky(data)
sky = f"Sky{sky_no}"
halo_data = true_halos.query("SkyId == @sky")
for i in range(halo_data["numberHalos"].item()):
    ax.scatter(halo_data[f"halo_x{i + 1}"], halo_data[f"halo_y{i + 1}"], c="k", s=60, label="True halo pos")
ax.legend()
```

## 

  
$$x_i \sim Uniform(0, 4200)$$  
$$y_i \sim Uniform(0, 4200)$$  
  
40-180  
$$m_{large} \sim \log Uniform(40, 180)$$  
  
$d_{i,j}$ $j$  $i$ 
$$e_i | (x,y) \sim Normal(\sum_{j=halos} d_{i,j}m_jf(r_{i,j}), \sigma ^2)$$
$f(r_{i,j})$ $j$  $i$   
<br>  
$$f(r_{i,j})=\frac{1}{max(r_{i,j}, 240)}$$
<br>:  
$$f(r_{i,j})=\frac{1}{max(r_{i,j}, 70)}$$

```python
def euclidean_distance(x: np.ndarray, y: np.ndarray):
    return np.sqrt(((x - y) ** 2).sum(axis=1))

def f_distance(galaxy_pos, halo_pos, c):
    return 1 / np.maximum(euclidean_distance(galaxy_pos, halo_pos), c).reshape(-1, 1)

def tangential_distance(galaxy_pos, halo_pos):
    delta = galaxy_pos - halo_pos
    t = 2 * np.arctan2(delta[:, 1], delta[:, 0]).reshape(-1, 1)
    return np.hstack([-np.cos(t), -np.sin(t)])
```

```python
mass_large = pm.Uniform("mass_large", 40, 180)
halo_position = pm.Uniform("halo_position", 0, 4200, size=(1, 2))


@pm.deterministic
def mean(mass=mass_large, halo_pos=halo_position, galaxy_pos=data[["x", "y"]].values):
    return mass * f_distance(galaxy_pos, halo_pos, 240) * tangential_distance(galaxy_pos, halo_pos)

ellpty = pm.Normal("ellipcity", mean, 20, observed=True, value=data[["e1", "e2"]].values)
```

```python
model = pm.Model([ellpty, mean, halo_position, mass_large])
map_ = pm.MAP(model)
map_.fit()
mcmc = pm.MCMC(model)
mcmc.sample(2e+5, 1.4e+5, 3)
```

```python
t = mcmc.trace("halo_position")[:].reshape(-1, 2)

ax = draw_sky(data, true_halos, sky_no)
ax.legend()
ax.scatter(t[:, 0], t[:, 1], alpha=0.015, c="r", label="possible halo pos")
```

```python
sky_no = 215
data = load_sky(sky_no)
data[:5]
```

```python
def halo_posteriors(n_halos, galaxy_data, samples=5e+5, burn_samples=34e+4, thin=4):
    mass_large = pm.Uniform("mass_large", 40, 180)
    mass_small_1 = 20
    mass_small_2 = 20
    
    masses = np.array([mass_large, mass_small_1, mass_small_2], dtype="object")
    
    halo_positions = pm.Uniform("halo_positions", 0, 4200, size=(n_halos, 2))
    fdist_constant = [240, 70, 70]
    
    @pm.deterministic
    def mean(masses=masses, halo_positions=halo_positions, glx_pos=galaxy_data[["x", "y"]].values, n_halos=n_halos):
        _sum = 0
        for i in range(n_halos):
            influence = masses[i] * f_distance(glx_pos, halo_positions[i, :], fdist_constant[i]) * tangential_distance(glx_pos, halo_positions[i, :])
            _sum += influence
        return _sum
    
    ellpty = pm.Normal("ellipcity", mean, 20, observed=True, value=galaxy_data[["e1", "e2"]].values)
    
    map_ = pm.MAP([ellpty, mean, halo_positions, mass_large])
    map_.fit(method="fmin_powell")
    
    mcmc = pm.MCMC([ellpty, mean, halo_positions, mass_large])
    mcmc.sample(samples, burn_samples, thin)
    return mcmc.trace("halo_positions")[:]
```

```python
traces = halo_posteriors(3, data, 10.5e+5, 9.5e+5, 10)
```

```python
ax = draw_sky(data)

colors = ["#467821", "#A60628", "#7A68A6"]
for i in range(traces.shape[1]):
    # possible distribution
    ax.scatter(traces[:, i, 0], traces[:, i, 1], c=colors[i], alpha=0.2)

sky = f"Sky{sky_no}"
halo_data = true_halos.query("SkyId == @sky")
for i in range(halo_data["numberHalos"].item()):
    ax.scatter(halo_data[f"halo_x{i + 1}"], halo_data[f"halo_y{i + 1}"], c="k", s=90, label="True halo pos")
ax.legend()
```
